```python
accs = [x[0] for x in task_1_datas]
precisions = [x[1] for x in task_1_datas]
recalls = [x[2] for x in task_1_datas]
aucs = [x[3] for x in task_1_datas]
```

```python
# Plot the accuracy
plt.figure(figsize=(10, 5))
plt.title("Accuracy")
plt.bar(task1_names, accs)
plt.ylim(0.8, 0.95)
plt.xticks(rotation=30)
for each in task1_names:
    plt.text(each, accs[task1_names.index(each)] + 0.001, str(round(accs[task1_names.index(each)], 3)), ha='center')
plt.show()
```

```python
# Plot the precision
plt.figure(figsize=(10, 5))
plt.title("Precision")
plt.bar(task1_names, precisions)
plt.ylim(0, 1)
plt.xticks(rotation=30)
for each in task1_names:
    plt.text(each, precisions[task1_names.index(each)] + 0.001, str(round(precisions[task1_names.index(each)], 3)), ha='center')
plt.show()
```

```python
# Plot the recall
plt.figure(figsize=(10, 5))
plt.title("Recall")
plt.bar(task1_names, recalls)
plt.ylim(0, 0.7)
plt.xticks(rotation=30)
for each in task1_names:
    plt.text(each, recalls[task1_names.index(each)] + 0.001, str(round(recalls[task1_names.index(each)], 3)), ha='center')
plt.show()
```

```python
# Plot the AUC
plt.figure(figsize=(10, 5))
plt.title("AUC")
plt.bar(task1_names, aucs)
plt.ylim(0.5, 1)
plt.xticks(rotation=30)
for each in task1_names:
    plt.text(each, aucs[task1_names.index(each)] + 0.001, str(round(aucs[task1_names.index(each)], 3)), ha='center')
plt.show()
```

```python
# Task 2
task2_names = ["Linear Regression",
                "Ridge Regression",
                "Lasso Regression",
                "Elastic Net",
                "FNN-6",
                "FNN-6 restricted",
                "KNN-10",
                "KNN-1000",
                "Random Forest",
                "Gradient Boosting",
                "Ada Boost",
                "Extra Trees",
                "Bagging",
                "XGBoost",
]

task_2_datas = [   [4.2, 9197.378572],
                    [3.4, 173.363853],
                    [2.5, 1.824610],
                    [5.7, 1.816818],
                    [30, 1047.13123],
                    [30, 2.02331],
                    [0.08, 2.057],
                    [0.08, 1.994],
                    [81, 1.747301],
                    [62, 1.761030],
                    [3.5, 1.962370],
                    [6.3, 1.753345],
                    [66, 1.750438],
                    [10.2, 1.739007],
                
                ]
```

```python
times = [x[0] for x in task_2_datas]
rmses = [x[1] for x in task_2_datas]
```

```python
# Plot the time
plt.figure(figsize=(10, 5))
plt.title("Time cost")
plt.bar(task2_names, times)
plt.ylim(0, 85)
plt.xticks(rotation=30)
for each in task2_names:
    plt.text(each, times[task2_names.index(each)] + 0.001, str(round(times[task2_names.index(each)], 3)), ha='center')
plt.show()
```

```python
# Plot the RMSE
plt.figure(figsize=(10, 5))
plt.title("RMSE")
plt.bar(task2_names, rmses)
plt.ylim(0, 10)
plt.xticks(rotation=30)
for i in range(len(task2_names)):
    if i == 0 or i == 1 or i == 4:
        # text at top as out of the bar
        plt.text(task2_names[i], 10 + 0.001, str(round(rmses[i], 3)), ha='center')
    else:
        # text at bottom as out of the bar
        plt.text(task2_names[i], rmses[i] - 0.001, str(round(rmses[i], 3)), ha='center')
plt.show()
```
